
## **Detección de Fraudes Financieros:** Desarrolla un sistema que utilice árboles de decisión para identificar transacciones financieras fraudulentas, considerando múltiples factores como patrones de gasto,tiempo,etc.

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import metrics
import pandas as pd
import os
import matplotlib.pyplot as plt
import random
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
import tensorflow as tf
%matplotlib inline
```

```python
path = os.getcwd() + '/sample_data/RegistroTarjeta1_2023.csv'
print(path)
Registros= pd.read_csv(path)
Registros = Registros.dropna()
Registros
```

```python
Registros.describe()
```

```python
# Organizacion de las columnas
Registros.columns = ['Time'] + [f'V{i}' for i in range(1, 29)] + ['Amount','Class']

#Creacion DataFrame X , Y

# Utiliza las columnas 'Time', 'Amount' y 'V1' a 'V28' como características (X)
X = Registros[['Time', 'Amount'] + [f'V{i}' for i in range(1, 29)]]

# La columna 'Class' como variable objetivo (y)
y = Registros['Class']
```

```python
# Divide los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Crea el clasificador de árbol de decisión
tree_classifier = DecisionTreeClassifier(max_features=None)

# Entrena el modelo
tree_classifier.fit(X_train, y_train)
```

```python
# Realiza predicciones en el conjunto de prueba
y_pred = tree_classifier.predict(X_test)
len(y_pred)
```

```python
# Muestra la matriz de confusión
metricas = metrics.confusion_matrix(y_test, y_pred)
print("Matriz de confusión:")
print(metricas)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = metricas, display_labels = ['No Fraude', 'Fraude'])

cm_display.plot()
plt.show()
```

```python
# Calcula el área bajo la curva de recuperación de precisión (AUPRC)
auprc = metrics.average_precision_score(y_test, y_pred)
print("Área bajo la curva de recuperación de precisión (AUPRC):", auprc)
```

```python
# Visualiza el árbol de decisión
plt.figure(figsize=(12, 6))
plot_tree(tree_classifier, filled=True, feature_names=X.columns, class_names=['Not Fraud', 'Fraud'])
plt.show()
```

```python
# Muestra la precisión del modelo
accuracy = metrics.accuracy_score(y_test, y_pred)
print("Precisión:", accuracy)
```

```python
# Crea el clasificador de árbol de decisión
tree_classifier1 = DecisionTreeClassifier(max_depth=3,max_features=None)

# Entrena el modelo
tree_classifier1.fit(X_train, y_train)

# Visualiza el árbol de decisión
plt.figure(figsize=(12, 6))
plot_tree(tree_classifier1, filled=True, feature_names=X.columns, class_names=['Not Fraud', 'Fraud'])
plt.show()
```

```python
# Generar valores aleatorios para V1 a V28
valores_aleatorios = [random.uniform(-3, 9) for _ in range(28)]

# Realiza una predicción de ejemplo
ejemplo = [[2, 1] + valores_aleatorios]  # Puedes cambiar los valores de 'Time', 'Amount', 'V1' a 'V28' aquí
prediccion = tree_classifier.predict(ejemplo)
print("Predicción del ejemplo:", "Fraud" if prediccion[0] == 1 else "Not Fraud")
print(prediccion[0])
```

```python
# Realiza una predicción de ejemplo para no fraude
ejemplo_fraude = [[2, 1000] + [1.0] * 28]
prediccion = tree_classifier.predict(ejemplo_fraude)
print("Predicción del ejemplo de fraude:", "Fraud" if prediccion[0] == 1 else "Not Fraud")
print(prediccion[0])
```

```python
# Realiza una predicción de ejemplo para un fraude
ejemplo_fraude = [[50, 50] + [-4.5] * 28]
prediccion = tree_classifier.predict(ejemplo_fraude)
print("Predicción del ejemplo de fraude:", "Fraud" if prediccion[0] == 1 else "Not Fraud")
print(prediccion[0])
```

### **RED NEURONAL**

```python
Registros.plot(kind='scatter', x='Amount', y='Class', figsize=(8,5))
```

```python
# Seleccionar características (X) y variable objetivo (y)
X1 = Registros[['Time', 'Amount'] + [f'V{i}' for i in range(1, 29)]]
y1 = Registros['Class']

# Crear un vector de características para la red neuronal
X_vector = X1.values
```

```python
# Crear el modelo de red neuronal
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=1, input_dim=X_vector.shape[1], activation='sigmoid'))

model.summary()
model.compile(optimizer=tf.keras.optimizers.Adam(0.001), loss='binary_crossentropy', metrics=['accuracy'])


# Entrenar el modelo
epochs_hist = model.fit(X_vector, y1, epochs=100, validation_split=0.2)
```

```python
epochs_hist.history.keys()
plt.plot(epochs_hist.history['loss'])
```

```python
# Visualizar el historial de pérdida
plt.plot(epochs_hist.history['loss'], label='Training Loss')
plt.plot(epochs_hist.history['val_loss'], label='Validation Loss')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

```python
# predicción de ejemplo
model.get_weights()
example_data = np.array([[90, 5000] + [random.uniform(-5, 9) for _ in range(28)]])
prediction1 = model.predict(example_data)
print("Predicción del ejemplo:", "Fraud" if prediction1[0][0] > 0.5 else "Not Fraud")
```

```python
# predicción de ejemplo para un fraude
model.get_weights()
ejemplo_fraude = [[9, 5] + [-4.5] * 28]
prediccion1 =  model.predict(ejemplo_fraude)
print("Predicción del ejemplo de fraude:", "Fraud" if prediccion1[0][0] > 0.5 else "Not Fraud")
```
